Try building the load profile from the parquet files.

```python
natural_gas_end_uses = [
    'out.natural_gas.heating.energy_consumption',
    'out.natural_gas.hot_water.energy_consumption',
    'out.natural_gas.range_oven.energy_consumption',
    'out.natural_gas.clothes_dryer.energy_consumption',
    'out.natural_gas.fireplace.energy_consumption',
]

kwh_to_therms = 0.0341296  # Conversion factor from kWh to therms

gas_8760_data = pd.DataFrame()

scenario = "scenario-baseline"
housing_type = "single-family-detached"
directory = f"alameda/{scenario}/{housing_type}/step1_buildings/"

# Counter for the number of buildings
building_count = 0

for file_name in os.listdir(directory):
    file_path = os.path.join(directory, file_name)
    if file_path.endswith('.parquet'):
        data = pd.read_parquet(file_path)
        data = data.reset_index()  # Reset the index to turn 'timestamp' into a column

        if 'timestamp' in data.columns and all(col in data.columns for col in natural_gas_end_uses):
            data['timestamp'] = pd.to_datetime(data['timestamp'])
            data = data[['timestamp'] + natural_gas_end_uses]
            data['calculated.natural_gas.total.kwh'] = data[natural_gas_end_uses].sum(axis=1)
            gas_8760_data = pd.concat([gas_8760_data, data], axis=0, ignore_index=True)
            building_count += 1  # Increment the building count

# Sum the values across all buildings
natural_gas_sum = gas_8760_data[natural_gas_end_uses].sum()

# Divide by the number of buildings to normalize the data
natural_gas_avg_per_building = natural_gas_sum / building_count

# Add the normalized total column in kWh
gas_8760_data['calculated.natural_gas.total.kwh'] = gas_8760_data[natural_gas_end_uses].sum(axis=1) / building_count

# Create a column for total therms
gas_8760_data['calculated.natural_gas.total.therms'] = gas_8760_data['calculated.natural_gas.total.kwh'] * kwh_to_therms

# Normalize the total annual natural gas consumption across all buildings
annual_natural_gas_used_by_all_buildings_kwh = gas_8760_data['calculated.natural_gas.total.kwh'].sum()
annual_natural_gas_used_by_all_buildings_therms = annual_natural_gas_used_by_all_buildings_kwh * kwh_to_therms

# Update the annual natural gas for each end use based on the number of buildings
for end_use in natural_gas_end_uses:
    gas_8760_data[end_use] = gas_8760_data[end_use] / building_count

print(annual_natural_gas_used_by_all_buildings_kwh)
print(annual_natural_gas_used_by_all_buildings_therms)
```

```python
scenario = 'scenario-baseline'
housing_type = "single-family-detached"
directory = f"alameda/{scenario}"

suitable_buildings_df = pd.read_csv(f"{directory}/{housing_type}/step0_out_suitable_buildings.csv")
building_ids = suitable_buildings_df['bldg_id'].unique().tolist()

metadata_file = f"{directory}/step0_in_CA_baseline_metadata_and_annual_results.csv"
building_metadata = pd.read_csv(metadata_file)
metadata_filtered = building_metadata[building_metadata['bldg_id'].isin(building_ids)]

natural_gas_end_use_columns = [
    'out.natural_gas.heating.energy_consumption.kwh',
    'out.natural_gas.hot_water.energy_consumption.kwh',
    'out.natural_gas.range_oven.energy_consumption.kwh',
    'out.natural_gas.clothes_dryer.energy_consumption.kwh',
    # 'out.natural_gas.fireplace.energy_consumption.kwh',
]

metadata_filtered['total_natural_gas_consumption_kwh'] = metadata_filtered[natural_gas_end_use_columns].sum(axis=1)

kwh_to_therms = 0.0341296
metadata_filtered['total_natural_gas_consumption_therms'] = metadata_filtered['total_natural_gas_consumption_kwh'] * kwh_to_therms

average_annual_usage_therms = metadata_filtered['total_natural_gas_consumption_therms'].mean()
print(f"\nAverage annual natural gas usage across all buildings (therms): {average_annual_usage_therms:.2f}")
print("\nTotal annual natural gas usage per building (therms):")
print(metadata_filtered[['bldg_id', 'total_natural_gas_consumption_therms']])

typical_annual_usage_therms = 400

def usage_comparison(usage):
    if usage < typical_annual_usage_therms * 0.5:
        return "Significantly lower than typical"
    elif usage > typical_annual_usage_therms * 1.5:
        return "Significantly higher than typical"
    else:
        return "Within expected range"

metadata_filtered['usage_comparison'] = metadata_filtered['total_natural_gas_consumption_therms'].apply(usage_comparison)
print("\nUsage comparison for each building:")
print(metadata_filtered[['bldg_id', 'total_natural_gas_consumption_therms', 'usage_comparison']])
```
