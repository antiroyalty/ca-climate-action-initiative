```python
import os
import pandas as pd

natural_gas_end_uses = [
    'out.natural_gas.heating.energy_consumption',
    'out.natural_gas.hot_water.energy_consumption',
    'out.natural_gas.range_oven.energy_consumption',
    'out.natural_gas.clothes_dryer.energy_consumption',
    'out.natural_gas.fireplace.energy_consumption',
    # 'out.natural_gas.total.energy_consumption'
]

all_data = pd.DataFrame()

for file_name in os.listdir(directory):
    file_path = os.path.join(directory, file_name)
    if file_path.endswith('.parquet'):
        data = pd.read_parquet(file_path)
        
        if 'timestamp' in data.columns and all(col in data.columns for col in natural_gas_end_uses):
            data['timestamp'] = pd.to_datetime(data['timestamp'])
            data = data[['timestamp'] + natural_gas_end_uses]
            
            # Add a unique building identifier
            data['building_id'] = file_name  # or extract an ID from the file name
            
            # Calculate total load per building per timestamp
            data['total_load'] = data[natural_gas_end_uses].sum(axis=1)
            
            all_data = pd.concat([all_data, data], axis=0, ignore_index=True)

# Calculate the mean total load across all buildings grouped by timestamp
average_profile = all_data.groupby('timestamp')['total_load'].mean()

# Create a date range covering the data range (8760 hours for a full year)
full_year = pd.date_range(
    start=all_data['timestamp'].min(), 
    end=all_data['timestamp'].max(), 
    freq='H'
)

# Reindex to ensure continuous hourly data and fill missing values with zeros
average_profile = average_profile.reindex(full_year).fillna(0)

# Convert Series to DataFrame
average_profile = average_profile.reset_index()
average_profile.columns = ['timestamp', 'average_total_load_kwh']

# Conversion factor from kWh to therms
kwh_to_therms = 0.0341296
average_profile['average_total_load_therms'] = average_profile['average_total_load_kwh'] * kwh_to_therms
```

```python
average_profile.to_csv(f"alameda/{scenario}/{housing_type}/step4_natural_gas_loads.csv", index=False)

average_profile.head()
```

```python
# Calculate the total annual natural gas usage in therms
total_annual_usage_therms = average_profile['average_total_load_therms'].sum()

print(f"Total annual natural gas usage per household (therms): {total_annual_usage_therms:.2f}")

# Ana Annual
# 12 + 13 + 27 + 62 + 47 + 51 + 35 + 27 + 10 + 11 + 8 + 10 = 313
# Typical annual usage for a California household (therms)
typical_annual_usage_therms = 400  # Adjust based on the latest data if necessary

# Validate whether the calculated usage is within a reasonable range
if total_annual_usage_therms < typical_annual_usage_therms * 0.5:
    print("The calculated annual usage is significantly **lower** than the typical usage for a California household.")
elif total_annual_usage_therms > typical_annual_usage_therms * 1.5:
    print("The calculated annual usage is significantly **higher** than the typical usage for a California household.")
else:
    print("The calculated annual usage is within the expected range for a California household.")
```

```python
natural_gas_end_uses = [
    'out.natural_gas.heating.energy_consumption',
    'out.natural_gas.hot_water.energy_consumption',
    'out.natural_gas.range_oven.energy_consumption',
    'out.natural_gas.clothes_dryer.energy_consumption',
    # 'out.natural_gas.fireplace.energy_consumption',
    # Include other end uses if available
    # 'out.natural_gas.grill.energy_consumption',
    # 'out.natural_gas.pool_heater.energy_consumption',
    # 'out.natural_gas.permanent_spa_heat.energy_consumption',
    # 'out.natural_gas.total.energy_consumption'
]

print(filepath)

building_metadata = pd.read_csv(f"alameda/{scenario}/step0_in_CA_baseline_metadata_and_annual_results.csv")
building_2917 = building_metadata[building_metadata['bldg_id'] == 2917]

building_2917['out.natural_gas.heating.energy_consumption.kwh']
building_2917['out.natural_gas.fireplace.energy_consumption.kwh']
building_2917['out.natural_gas.clothes_dryer.energy_consumption.kwh']
building_2917['out.natural_gas.hot_water.energy_consumption.kwh']

data = pd.read_parquet(filepath) # one specific parquet file

if 'timestamp' in data.columns and all(col in data.columns for col in natural_gas_end_uses):
    data['timestamp'] = pd.to_datetime(data['timestamp'])
    
    data = data[['timestamp'] + natural_gas_end_uses]
    
    data['total_load_kwh'] = data[natural_gas_end_uses].sum(axis=1)
    
    # Set 'timestamp' as the index
    data.set_index('timestamp', inplace=True)
    
    # Create a date range covering the full year (8760 hours)
    full_year = pd.date_range(
        start=data.index.min(), 
        end=data.index.max(), 
        freq='H'
    )
    
    data = data.reindex(full_year).fillna(0)
    data.index.name = 'timestamp'  # Ensure the index is named 'timestamp'
    # Reset the index to turn it back into a column
    data = data.reset_index()
    
    # Conversion factor from kWh to therms
    conversion = 0.01 
    # Assume: units are kbtu, need to divide by 100 to get to therms
    # Assume: units are in kwh, 1 kWh â‰ˆ 0.0341296 therms
    
    data['total_load_therms'] = data['total_load_kwh'] * conversion
    
    # data.to_csv("step4_natural_gas_loads_single_building.csv", index=False)
    
    data.head()
else:
    print("Required columns are missing in the data.")


# Calculate the total annual natural gas usage in therms
total_annual_usage_therms = data['total_load_therms'].sum()

print(f"\nTotal annual natural gas usage for the building (therms): {total_annual_usage_therms:.2f}")

# Typical annual usage for a California household (therms)
typical_annual_usage_therms = 400

if total_annual_usage_therms < typical_annual_usage_therms * 0.5:
    print("The calculated annual usage is significantly **lower** than the typical usage for a California household.")
elif total_annual_usage_therms > typical_annual_usage_therms * 1.5:
    print("The calculated annual usage is significantly **higher** than the typical usage for a California household.")
else:
    print("The calculated annual usage is within the expected range for a California household.")
```
