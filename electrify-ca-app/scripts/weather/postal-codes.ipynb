```python
postal_code_tsv = [row.split('\t') for row in open("./geonames-ca-postal-codes.txt").readlines()]

values = [(r[1], r[2], r[4], float(r[-3]), float(r[-2])) for r in postal_code_tsv]

values
```

```python
postal_dict = {}
for (postal_code, name, province_code, lat, lng) in values:
    postal_dict[postal_code] = {"placeName": name, "provinceCode": province_code, "lat": lat, "lng": lng}

len(postal_dict)
```

```python
import json

with open('ca-postal-codes.json', 'w') as file:
    json.dump(postal_dict, file)
```
